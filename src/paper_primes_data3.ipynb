```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def draw_figure():
    # Set up the plot
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Set background to transparent
    fig.patch.set_alpha(0)
    ax.patch.set_alpha(0)

    # Hide axes
    # Adjusted x_lim to make room for labels on the left
    ax.set_xlim(-3.5, 13.5)
    ax.set_ylim(0, 4)
    ax.axis('off')

    # Define styles
    line_color = '#5f8c9e'
    dot_color = 'black'
    font_family = 'serif'
    tick_fontsize = 14
    
    # ---------------------------------------------------------
    # Row 1: The original sequence (All L)
    # ---------------------------------------------------------
    y1 = 2.5
    
    # Add Label N on the left
    ax.text(-0.5, y1, 'N', fontsize=16, va='center', ha='right', fontfamily=font_family)
    
    # Axis line
    ax.arrow(0, y1, 12.8, 0, head_width=0.1, head_length=0.2, fc=line_color, ec=line_color)
    
    for i in range(13):
        # Always solid dot for top row
        circle = plt.Circle((i, y1), 0.1, color=dot_color, zorder=3)
        ax.add_patch(circle)
        # Label L(i)
        ax.text(i, y1 - 0.3, f'L({i})', ha='center', fontsize=tick_fontsize, fontfamily=font_family)

    # ---------------------------------------------------------
    # Row 2: The composite sequence M2  M3
    # ---------------------------------------------------------
    y2 = 1.0
    
    # Add Label M2  M3 on the left
    ax.text(-0.5, y2, r'$M_2 \bullet M_3$', fontsize=16, va='center', ha='right', fontfamily=font_family)
    
    # Axis line
    ax.arrow(0, y2, 12.8, 0, head_width=0.1, head_length=0.2, fc=line_color, ec=line_color)
    
    # Logic for R points: Divisible by 2 OR 3
    for i in range(13):
        is_R = (i % 2 == 0) or (i % 3 == 0)
        
        if is_R:
            # Hollow circle for R
            circle = plt.Circle((i, y2), 0.1, facecolor='white', edgecolor=dot_color, zorder=3)
            ax.add_patch(circle)
            # Label R(i)
            ax.text(i, y2 - 0.3, f'R({i})', ha='center', fontsize=tick_fontsize, fontfamily=font_family)
        else:
            # Solid dot for L
            circle = plt.Circle((i, y2), 0.1, color=dot_color, zorder=3)
            ax.add_patch(circle)
            # Label L(i)
            ax.text(i, y2 - 0.3, f'L({i})', ha='center', fontsize=tick_fontsize, fontfamily=font_family)

    # ---------------------------------------------------------
    # Draw Arrows for Period 2 and Period 3 on the bottom row
    # ---------------------------------------------------------
    # Period 2 arrows (connects even numbers)
    for i in range(0, 11, 2):
        # Slightly higher arc for period 2
        arrow = patches.FancyArrowPatch((i, y2+0.15), (i+2, y2+0.15), 
                                      connectionstyle="arc3,rad=-0.5", 
                                      color='black', 
                                      arrowstyle='-|>',
                                      mutation_scale=12,
                                      linestyle='--',
                                      linewidth=0.5,
                                      alpha=0.7)
        ax.add_patch(arrow)

    # Period 3 arrows (connects multiples of 3)
    for i in range(0, 10, 3):
        # Slightly flatter/different arc for period 3 to distinguish or overlay
        # In the original image, they overlap. We use a slightly different rad or just layer them.
        arrow = patches.FancyArrowPatch((i, y2+0.15), (i+3, y2+0.15), 
                                      connectionstyle="arc3,rad=-0.6", 
                                      color='black', 
                                      arrowstyle='-|>',
                                      mutation_scale=12,
                                      linestyle='--',
                                      linewidth=0.5,
                                      alpha=0.7)
        ax.add_patch(arrow)

    # ---------------------------------------------------------
    # Bottom Text Formula
    # ---------------------------------------------------------
    # Using mathtext for the formula M_2  M_3 = RLRRRL
    ax.text(6, 0.2, r'$M_2 \bullet M_3 = RLRRRL$', ha='center', fontsize=16, fontfamily=font_family)

    plt.tight_layout()
    plt.savefig('prime_period_figure_4.png', dpi=300, bbox_inches='tight', transparent=True)
    print("Image saved to prime_period_figure_4.png")

if __name__ == "__main__":
    draw_figure()
```
